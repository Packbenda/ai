Hello this is a blood/ketchup classifier. If you don't know if it's ketchup or blood you can take a picture and upload it here, the classifier will tell you if it's blood or ketchup.
ps: don't eat it if you are hesitating.

```python
path = Path()
learn_inf = load_learner(path/"export.pkl", cpu=True)

def on_click_classify(change):
  img = PILImage.create(uploader.data[0])
  with out_pl: display(img.to_thumb(256,256))
  what_is,_, probs = learn_inf.predict(img)
  print(f"It is: {what_is}.")

btn_run = widgets.Button(description='Classify')
uploader = widgets.FileUpload()
btn_run.on_click(on_click_classify)
out_pl = widgets.Output()


VBox([uploader, btn_run, out_pl])
```
